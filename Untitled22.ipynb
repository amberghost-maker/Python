```python
import cv2

# Corrected typo: haarcasdes -> haarcascades and .xtmml -> .xml
face_cascde = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Initialize VideoCapture object, typically with an integer for camera index (e.g., 0)
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open camera.")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: Failed to capture image")
        break

    # Corrected typos: ctColour -> cvtColor and COLOUR_BGR2GRAY -> COLOR_BGR2GRAY
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = face_cascde.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5,
                                         minSize=(30,30))
    for (x, y, w, h) in faces:
        cv2.rectangle(frame,(x,y), (x + w, y+h), (255, 0, 0), 2) # Added color tuple for rectangle

    cv2.imshow('Face Detection - Press q to Quit',frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
