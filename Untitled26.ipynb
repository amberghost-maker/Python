```python
import pygame
import sys

pygame.init()

screen = pygame.display.set_mode((500, 500))

pygame.display.set_caption("My first game screen")

samurai_image = pygame.image.load("samurai.jpg")

samurai_image = pygame.transform.scale(samurai_image, (300, 300))

background_color = (58, 58, 58)

running = True
while running:

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    screen.fill(background_color)

    image_x = (500 - 300) // 2   # 100
    image_y = (500 - 300) // 2   # 100

    screen.blit(samurai_image, (image_x, image_y))

    pygame.display.flip()

pygame.quit()
sys.exit()
```
